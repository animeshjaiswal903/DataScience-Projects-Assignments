```python
#Data ready for TRAIN DATA
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

df = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
df['Survived'] = df['Survived'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')
df['Embarked'] = df['Embarked'].astype('category')
df['Sex'] = df['Sex'].astype('category')
df['Parch'] = df['Parch'].astype('category')
df['SibSp'] = df['SibSp'].astype('category')

#feature extraction (extra steps for knowledge)
#df['Title'] = df['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
#df['Age'] = df['Age'].fillna(df.groupby('Title')['Age'].transform('mean'))
#df['Age'] = df['Age'].fillna(df.groupby('Sex')['Age'].transform('mean'))
#df.drop(["Name","Ticket","Cabin","PassengerId","Title"],axis=1,inplace=True)

df.drop(["Name","Ticket","Cabin","PassengerId"],axis=1,inplace=True)
df.dtypes



dummy=pd.get_dummies(df[["Sex","Pclass","Embarked","Parch","SibSp"]])   #it creates a table of column headers as the values inside the column fuel-type and its values are 1,0
df=pd.concat([df,dummy],axis=1)   #merges the two dataframes along columns

#Filling NaN
mfa=df["Age"][df["Sex_female"]==1].mean()
mma=df["Age"][df["Sex_male"]==1].mean()
df['Age'][df["Sex_female"]==1]=df['Age'][df["Sex_female"]==1].fillna(round(mfa,0))
df['Age'][df["Sex_male"]==1]=df['Age'][df["Sex_male"]==1].fillna(round(mma,0))

df["Embarked"].fillna(df["Embarked"].mode()[0],inplace=True)
df.drop(["Sex","Pclass","Embarked","Parch","SibSp"],axis=1,inplace=True)

#Model building for TRAIN data
from sklearn.linear_model import LogisticRegression as LogReg
from sklearn.metrics import f1_score
x = df.drop(['Survived'], axis=1)
y = df['Survived']
df.head()

from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(x,y,test_size=0.2, random_state = 57)

#scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
cols = train_x.columns
train_x_scaled = scaler.fit_transform(train_x)
train_x_scaled = pd.DataFrame(train_x_scaled, columns=cols)
train_x_scaled.head()
test_x_scaled = scaler.transform(test_x)
test_x_scaled = pd.DataFrame(test_x_scaled, columns=cols)
test_x_scaled.head()
```

```python
#LogReg model

logreg = LogReg()
# Fitting the model
logreg.fit(train_x_scaled, train_y)
train_predict = logreg.predict(train_x_scaled)
train_k = f1_score(train_predict, train_y)
print("Train f1_score:",train_k)
test_predict = logreg.predict(test_x_scaled)
k = f1_score(test_predict, test_y)
print('Test f1_score:', k )

```
